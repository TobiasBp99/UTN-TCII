## Parte I

Podemos plantear dos cuadripolos representados por sus **_parámetros T interconectados en cascada._**
<img src="./cuadri.png" align="center" width="500" />
Por simple inspección se puede hallar los **_parámetros Z del cuadripolo izquierdo_** para posteriormente **_convertirlos_**. Además al conocer como **_afecta la carga de una impedancia a un sólo cuadripolo_** no es necesario trabajar con toda la matriz de parámetros.

\begin{equation}
\label{eqn:n1}
T = T_\alpha \cdot T_\beta
\end{equation}


### Cuadripolo $\alpha$

Hallamos los parámetros Z y luego los convertimos

$ZA = s \cdot L_1$

$Zb = \frac{1}{s \cdot C_2}$

$ZC = s \cdot L_3$

\begin{equation}
\begin{pmatrix}
ZA+ZB & ZB \\
ZB & ZC+ZB
\end{pmatrix}
\end{equation}

Operando con las ecuaciones de los parámetros podemos convertirlos a los parámetros deseados para buscar la transferencia

\begin{equation}
\label{eqn:n2}
V_1 = I_1 \cdot Z_{11} + I_2 \cdot Z_{12}
\end{equation}

\begin{equation}
\label{eqn:n3}
V_2 = I_1 \cdot Z_{21} + I_2 \cdot Z_{22}
\end{equation}

\begin{equation}
\label{eqn:n4}
A_\alpha = \frac{V_1}{V_2}|_{I_2 = 0} = \frac{Z_{11}}{Z_{21}}
\end{equation}

\begin{equation}
\label{eqn:n5}
B_\alpha = \frac{V_1}{-I_2}|_{V_2 = 0} = \frac{Z_{22}}{Z_{21}} - Z_{21}  
\end{equation}

### Cuadripolo $\beta$

Este cuadripolo también lo obtenemos por simple inspección

\begin{equation}
\begin{pmatrix}
1 & 0 \\
\frac{1}{R} & 1
\end{pmatrix}
\end{equation}

### Interconexión

Como sólo nos interesa la transferencia de tensión **_vamos a centrar el estudio únicamente en el parámetro A el cual es la ganancia inversa de tensión_**, al cargar un cuadripolo podemos absorver este efecto de la siguiente forma.

\begin{equation}
\label{eqn:n6}
A_T = A_\alpha + \frac{B_\alpha}{R_L} = \frac{s^{3} \cdot L_3 \cdot L_1 \cdot C_2 + s^{2} \cdot R \cdot L_1 \cdot C_2 + s \cdot (L_3 + L_1) + R }{R}   
\end{equation}

Simplemente invirtiendo la fracción de la ganancia inversa obtendremos la transferencia

\begin{equation}
\label{eqn:n7}
Tv_{(s)} = \frac{R}{s^{3} \cdot L_3 \cdot L_1 \cdot C_2 + s^{2} \cdot R \cdot L_1 \cdot C_2 + s \cdot (L_3 + L_1) + R}   
\end{equation}

\begin{equation}
\label{eqn:n8}
Tv_{(s)} = \frac{1}{s^{3} + s^{2} 2 + s 2 + 1}   
\end{equation}

Analizando vemos que esto es un filtro pasabajos de máxima planicidad.

## Parte III

### Brutton

La transformación de Brutton consiste en afectar a todas las impedancias del circuito al multiplicarlas por 's'. Obteniendo así una transformación a nivel de componentes.

$ YB_{(s)} = Y_{(s)} \cdot s $

### Construcción 

Para utilizar el FDNR se optó por adaptar un GIC para modelarlo. Utilizando dos capacitores y tres resistencias se puede obtener la función excitación necesaria. Además optando las Z2 y Z3 del mismo valor se puede demostrar que **_las pérdidas_** en el componente a emular son mucho menores.

Si se utiliza la transformada de Brutton se podrá obtener un cuadripolo equivalente, es decir que se comporta de la misma manera que el original, esto puede ser útil para evitar el uso de componentes como inductores que son difíciles de conseguir/realizar. En el caso de la aplicación de propuesta la resistencia modela un parlante que convierte una señal eléctrica en una onda mecánica, y toda la potencia desarrollada en dicho resistor es activa.

En el caso de la conversión la resistencia ahora debe ser reemplazada por un capacitor y la potencia en él tiene naturaleza reactiva que no podrá ser aprovehcada. La potencia activa se producirá en los resistores.
