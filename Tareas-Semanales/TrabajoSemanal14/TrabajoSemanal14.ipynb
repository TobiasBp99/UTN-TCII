## Consigna

Diseñe un atenuador de banda ancha que permita atenuar 30 dB intercalado en un cable coaxial de 75 ohms. 


a) Obtenga una red que satisfaga los requerimientos.
b) Verifique que el atenuador cumpla con la impedancia y atenuación prescrita.

### Diseño

Sabiendo que un atenuador lo podemos representar por su __*matriz T relacionada*__, gracias a la impedancia característica y el parámetro $\gamma$ plantemos

T = $\bigl(\begin{smallmatrix}\cosh{\gamma} & \sinh{\gamma} \cdot zo \\ \frac{\sinh{\gamma}}{zo} & cosh{\gamma}\end{smallmatrix}\bigr)$

Viendo que estamos en presencia de un atenuador podemos decir que $\gamma$ será real y para atenuar 30 dB $\alpha = \frac{30 dB}{\frac{8,69 dB}{nepper}} = 3,45 nepper $

Ya estamos en condiciones de __*otorgar valores a nuestra matriz T*__ y posteriormente podremos sintetizarlo con una red T o red $\phi\$ __*convirtiendo la matriz de parámetros a otro juego*__ que se condiga con la nueva familia.

T = $\bigl(\begin{smallmatrix}15,8 & 1182,73 \\ 0,21 & 15,8\end{smallmatrix}\bigr)$

Ahora convirtiendo la matriz a los parámetros impedancia obtenemos Z = $\bigl(\begin{smallmatrix}75,15 & 4,75 \\ 4,75 & 75,15\end{smallmatrix}\bigr)$

Fácilmente podemos observar que nuestra red es simétrica, es decir que manejamos un mismo nivel de impedancia en ambos puertos.

## Consigna

Diseñe un acoplador/atenuador de banda ancha que permita interconectar un cable coaxil de 75 Ω a otro de 50 Ω. Es decir, la red en su entrada deberá permitir la conexión una ZI1=75Ω y en su salida ZI2=50Ω de impedancia característica.

a) Obtenga una red que satisfaga los requerimientos y atenúe 5.72 dB en potencia.
b) Verifique que el atenuador cumpla con la impedancia y atenuación prescrita. 
c) ¿Podría dicha red acoplar sin atenuar? 
d) Si la atenuación requerida fuera de 80 dB, ¿cómo cambiaría la red calculada en a)? Proponga una solución en caso que no se pueda implementar.

### Diseño

Sabiendo que un atenuador lo podemos representar por su __*matriz T relacionada*__, considerando los niveles de impedancia a manejar y el parámetro $\gamma$ plantemos

T = $\bigl(\begin{smallmatrix}\cosh{\gamma} \cdot \sqrt{\frac{ZI_1}{ZI_2}} & \sinh{\gamma} \cdot \sqrt{ZI_1 \cdot ZI_2} \\ \frac{\sinh{\gamma}}{\sqrt{ZI_1 \cdot ZI_2}} & cosh{\gamma} \cdot \sqrt{\frac{ZI_2}{ZI_1}} \end{smallmatrix}\bigr)$

Viendo que estamos en presencia de un atenuador podemos decir que $\gamma$ será real y para atenuar 5,72 dB $\alpha = \frac{5,72 dB}{\frac{8,686 dB}{nepper}} = 0,6585 nepper $

Ya estamos en condiciones de __*otorgar valores a nuestra matriz T*__ y posteriormente podremos sintetizarlo con una red T o red $\phi\$ __*convirtiendo la matriz de parámetros a otro juego*__ que se condiga con la nueva familia.

T = $\bigl(\begin{smallmatrix}1,5 & 43,3 \\ 0,0115 & 1\end{smallmatrix}\bigr)$

Ahora convirtiendo la matriz a los parámetros impedancia obtenemos Z = $\bigl(\begin{smallmatrix}130,43 & 86,95 \\ 86,95 & 86,95\end{smallmatrix}\bigr)$

Hay varias cosas llamativas en esta matriz, por ejemplo la coincidencia en el valor numérico de los parámetros nos fuerza a que el componente en derivación sea el encargado de aportar el nivel de impedancia para realizar el acople.

Respecto a si es posible acoplar sin atenuar, en primer lugar al implementar el circuito con componentes resistivos habrá una cierta atenuación imposible de evitar. De hecho tomando el ejemplo realizado vemos que estamos en el límite de realizabilidad en nuestro circuito porque un resistor toma valor 0, entonces es un cable. Esto nos muestra que estamos frente a la atenuación mínima en la red propuesta.

El caso  contrario, si uno empezara a exigir mayor atenuación, habría que analizar si la realizabilidad no pasa a depender de otro componente. En caso de que no sea viable realizar el acoplador, se podría relajar el acoplador implementandolo en cascadas de la misma etapa.

Viendo que estamos en presencia de un atenuador podemos decir que $\gamma$ será real y para atenuar 80 dB $\alpha = \frac{80 dB}{\frac{8,686 dB}{nepper}} = 9,2070 nepper $

Ya estamos en condiciones de __*otorgar valores a nuestra matriz T*__ y posteriormente podremos sintetizarlo con una red T o red $\phi\$ __*convirtiendo la matriz de parámetros a otro juego*__ que se condiga con la nueva familia.

T = $\bigl(\begin{smallmatrix}6123 & 306150,3857 \\ 81,64 & 4082\end{smallmatrix}\bigr)$

Ahora convirtiendo la matriz a los parámetros impedancia obtenemos Z = $\bigl(\begin{smallmatrix}75 & 0,00427 \\ 0,0122 & 50\end{smallmatrix}\bigr)$

Rápidamente notamos que __*no podrá realizarse este circuito*__ con una resistencia de 0,0122 $\Omega$

El procedimiento para relajar el sistema será el mismo que venimos realizando, partiendo de la atenuación obtendremos la matriz T y luego su equivalente Z.
